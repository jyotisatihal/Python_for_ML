**Linear regression using x and y values**

```python
#sickit learn
#stats model

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x=np.array([1,5,13,25,33,46]).reshape((-1,1))
y=np.array([3,6,15,12,21,18])
plt.scatter(x,y)
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
SLR=LinearRegression()
SLR.fit(x,y)
pred=SLR.predict(x)

from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,pred)
plt.plot(x,pred,color="HotPink",marker="*")
print("MSE:",mse)
print("intercept:",SLR.intercept_)
print("slope:",SLR.coef_)
```

**RANDOMLY GENERATE X AND Y USING RANDOM MODULE AND DO LINEAR REGRESSION**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

np.random.seed(0)
x = np.random.randint(0, 50, size=(6,1))
y = np.random.randint(0, 30, size=6)

plt.scatter(x, y)
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()

SLR = LinearRegression()
SLR.fit(x, y)
pred = SLR.predict(x)

mse = mean_squared_error(y, pred)
plt.plot(x, pred, color="HotPink", marker="*")
print("MSE:", mse)
print("Intercept:", SLR.intercept_)
print("Slope:", SLR.coef_)
```
