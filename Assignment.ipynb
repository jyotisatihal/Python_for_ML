Calculate mean,median and mode

```python
import pandas as pd
iris=pd.read_csv('/content/iris.csv')
print(iris)
print("\nMean\n")
print(iris.mean(numeric_only=True))
print("\nMedian\n")
print(iris.median(numeric_only=True))
print("\nMode\n")
print(iris.mode(numeric_only=True))
```

Lottery game and success rate

```python
import random
import matplotlib.pyplot as plt
import numpy as np
comp=0
tie=0
user=0
for z in range(1,9):
 computer=random.randint(1,9)
 user=int(input("enter your number bw 1 to 9  "))
 print(computer)
 print(user)

 if computer>user:
  print("computer won",comp,"times")
  comp=comp+1
 elif computer == user:
  print("tied",tie,"times")
  tie=tie+1
 else:
  print("you won",user,"times")
  user=user+1

x=np.array(["user","computer","tie"])
y=np.array([user,comp,tie])
bar=plt.bar(x,y)
plt.title("success rate")
plt.ylabel("succeeded ")
bar[0].set_color('pink')
bar[1].set_color('blue')
bar[2].set_color('yellow')
plt.show()
```
